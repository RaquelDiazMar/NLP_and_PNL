### Reading text data into Python

**Read In Semi-structured Text Data**

The raw dataset being used in this course can be found at: http://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection

```python
# Read in and view the raw data
import pandas as pd

messages = pd.read_csv('spam.csv', encoding='latin-1')
messages.head()
```

```python
# Drop unused columns and label columns that will be used
messages = messages.drop(labels = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis = 1)
messages.columns = ["label", "text"]
messages.head()
```

```python
# is any data missing?
print('Number of nulls in label: {}'.format(messages['label'].isnull().sum())) # 'sum' function sums only the positive values, the ones which are null in this case
print('Number of nulls in text: {}'.format(messages['text'].isnull().sum()))
```

### Cleaning text data

1. Remove punctuation
2. Tokenization
3. Remove stopwords

#### Remove punctuation

```python
# What punctuation is included in the default list?
import string

string.punctuation
```

```python
def remove_punct(text):
    text = "".join([char for char in text if char not in string.punctuation])
    return text

messages['text_clean'] = messages['text'].apply(lambda x: remove_punct(x))

messages.head()
```

#### Tokenization

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

messages['text_tokenized'] = messages['text_clean'].apply(lambda x: tokenize(x.lower()))

messages.head()
```

```python
# anoher way to tokenize
def tokenize2(text):
    tokens2 = nltk.word_tokenize(text)
    return tokens2

messages['text_tokenized2'] = messages['text_clean'].apply(lambda x: tokenize2(x.lower()))

messages.head()
```

#### Remove stopwords

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
```

```python
print(len(stopwords))
```

```python
def remove_stopwords(tokenized_text):    
    text = [word for word in tokenized_text if word not in stopwords]
    return text

messages['text_nostop'] = messages['text_tokenized'].apply(lambda x: remove_stopwords(x))

messages.head()
```

### Vectorize text using Term Frequency - Inverse Document Frequency (TF-IDF)

How to convert text into a numeric representation to be passed into a machine learning model.
* Creates a document-term matrix; one row per document one column per word in the corpus.
* Generates a weighting for each word/document pair intended to reflect how important a given word is to the document within the context of its frequency within a larger corpus.



```python
# Fit a basic TFIDF Vectorizer 
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=remove_stopwords) # pass the function to clean your text
X_tfidf = tfidf_vect.fit_transform(messages['text_nostop'])# fit and transform the data in one matrix, a sparse matrix, a matrix in which most entries are zero
print(X_tfidf.shape, '\n')
print(tfidf_vect.get_feature_names()) # returns all the words that are vectorized, all the columns
```

```python
print(len(tfidf_vect.get_feature_names()))
```

```python
# How is the output of TfidfVectorizer stored?
X_tfidf
```

### Building a model on top of vectorized text

```python
X_features = pd.DataFrame(X_tfidf.toarray()) # convertin gthe sparse matrix to a data frame
X_features.head()
```

### Explore RandomForestClassifier Attributes & Hyperparameters

```python
# Import Random Forest for classification from sklearn
from sklearn.ensemble import RandomForestClassifier
```

```python
# arguments (and default values) for RandomForestClassifier
print(RandomForestClassifier())
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_features,
                                                    messages['label'],
                                                    test_size=0.2)
```

```python
# Fit a basic Random Forest model
rf = RandomForestClassifier()
rf_model = rf.fit(X_train, y_train)
```

```python
# Make predictions on the test set using the fit model
y_pred = rf_model.predict(X_test)
```

```python
# Evalute model predictions using precision and recall
precision = precision_score(y_test, y_pred, pos_label='spam')
recall = recall_score(y_test, y_pred, pos_label='spam')
print('Precision: {} / Recall: {}'.format(round(precision, 3), round(recall, 3)))
```

In the context of a spam filter:
* 100% precision means that when the model identified something as spam, it was actually spam 100% of the time.  

* 82.5% recall means that of all the spam that came into your email, 82.5% of it was properly placed in the spam folder, which means the other 17.5% went into your inbox.
